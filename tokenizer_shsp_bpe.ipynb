```python
raw = data.encode('utf-8')
tokens = list(map(int, raw))
len(tokens)
```

```python
def count_pairs(ids):
  counts = defaultdict(int)

  for pair in zip(ids, ids[1:]):
    counts[pair] += 1

  return counts

def replace(ids, pair, new_tok):
  newli = []
  i = 0
  while i < len(ids):
    if i < len(ids)-1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
      newli.append(new_tok)
      i += 2
    else:
      newli.append(ids[i])
      i+=1

  return newli
```

```python
# train loop
new_tok = 500
rdata = tokens
tokenizer = {}
```

```python
for i in range(new_tok):
  stat = count_pairs(rdata)
  pair = max(stat, key = lambda p: stat[p])
  tok = 256 + i
  re = replace(rdata, pair, tok)
  rdata = re
  tokenizer[pair] = tok
```

```python
tokenizer
```

```python
vocab = {i: bytes([i]) for i in range(256)}
for (p0, p1), tok in tokenizer.items():
  vocab[tok] = vocab[p0] + vocab[p1]
```

```python
vocab
```

```python
def encode(text):
  tokens = list(text.encode('utf-8'))
  while len(text) >= 2:
    get_stat = count_pairs(tokens) # retunes dict of pairs and counts
    # get the pair which is in the tokenizer from get stat at the min pos
    pairs = min(get_stat, key = lambda p:tokenizer[p] if p in tokenizer else float('inf')) #first mergable pair
    if pairs not in tokenizer:
      break
    tok = tokenizer[pairs]
    tokens = replace(tokens, pairs, tok)

  return tokens
```

```python
def decode(lis):
  raw = b''.join(vocab[i] for i in lis)
  return raw.decode('utf-8')
```
